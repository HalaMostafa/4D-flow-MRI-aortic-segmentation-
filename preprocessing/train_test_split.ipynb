```python
import os
import random
import shutil

BASE_DIR = os.path.dirname(os.path.abspath(os.getcwd()))
# print(BASE_DIR)
RAW_DATA_DIR = os.path.join(BASE_DIR, '/workspace/data')
IMAGES_TR = os.path.join(RAW_DATA_DIR, 'ao_imgs')
LABELS_TR = os.path.join(RAW_DATA_DIR, 'ao_segs')

# Output dirs
SAMPLE_DIR = os.path.join(BASE_DIR, '/workspace/sample_data')
TRAIN_DIR = os.path.join(BASE_DIR, '/workspace/train_data')

for d in [SAMPLE_DIR, TRAIN_DIR]:
    os.makedirs(os.path.join(d, 'ao_imgs'), exist_ok=True)
    os.makedirs(os.path.join(d, 'ao_segs'), exist_ok=True)

# ✅ Step 1: Build mapping by patient ID
pairs = {}

# Process images
for img_file in os.listdir(IMAGES_TR):
    if img_file.endswith('.nii.gz'):
        patient_id = img_file.replace('_0000.nii.gz', '')
        pairs[patient_id] = {
            "ao_imgs": img_file,
            "ao_segs": None
        }

# Process labels
for lbl_file in os.listdir(LABELS_TR):
    if lbl_file.endswith('.nii.gz'):
        patient_id = lbl_file.replace('.nii.gz', '')
        if patient_id in pairs:
            pairs[patient_id]["ao_segs"] = lbl_file

# ✅ Step 2: Keep only VALID pairs
valid_pairs = []
for pid, data in pairs.items():
    if data["ao_imgs"] and data["ao_segs"]:
        valid_pairs.append((pid, data["ao_imgs"], data["ao_segs"]))

print(f"Valid pairs: {len(valid_pairs)}")

# ✅ Step 3: Shuffle by patient ID
random.seed(42)
random.shuffle(valid_pairs)

# ✅ Step 4: Split by IDs (NOT files)
sample_ratio = 0.6
split_idx = int(len(valid_pairs) * sample_ratio)

sample_pairs = valid_pairs[:split_idx]
train_pairs = valid_pairs[split_idx:]

print(f"Sample patients: {len(sample_pairs)}")
print(f"Train patients: {len(train_pairs)}")

# ✅ Step 5: Copy function
def copy_pairs(pairs_list, dst_dir):
    for pid, img_file, label_file in pairs_list:
        shutil.copy(
            os.path.join(IMAGES_TR, img_file),
            os.path.join(dst_dir, 'ao_imgs', img_file)
        )
        shutil.copy(
            os.path.join(LABELS_TR, label_file),
            os.path.join(dst_dir, 'ao_segs', label_file)
        )

# Copy
copy_pairs(sample_pairs, SAMPLE_DIR)
copy_pairs(train_pairs, TRAIN_DIR)

print("Done ✅")
```
